```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

Aksiya narxini bashorat qilish
Qimmatli qog'ozlar bozorini bashorat qilish uning paydo bo'lishidan beri investorlarning muammosi va maqsadi bo'lib kelgan. Birjada har kuni milliardlab dollarlar ayirboshlanadi va har bir dollar ortida u yoki bu yoâ€˜l bilan foyda olishga umid qilayotgan investor turibdi.

Butun kompaniyalar bozor xatti-harakatlariga qarab har kuni ko'tariladi va tushadi. Agar investor bozor harakatlarini aniq bashorat qila olsa, u boylik va ta'sirning ajoyib va'dasini taklif qiladi. 


Bugungi kunda juda ko'p odamlar uyda o'tirib, fond bozorida savdo qilishmoqda. Agar siz qimmatli qog'ozlar bozoridagi tajribangiz va mashinani o'rganish ko'nikmalaringizdan qimmatli qog'ozlar narxini bashorat qilish uchun foydalansangiz, bu siz uchun ortiqcha nuqtadir.

Keling, Machine Learning va python dasturlash tilidan foydalangan holda aksiyalar narxini qanday bashorat qilishni ko'rib chiqaylik. Men ushbu vazifani bajarish uchun zarur bo'lgan barcha kerakli python kutubxonalarini import qilish orqali boshlayman:

```python
def prepare_data(df,forecast_col,forecast_out,test_size):
    label = df[forecast_col].shift(-forecast_out) #creating new column called label with the last 5 rows are nan
    X = np.array(df[[forecast_col]]) #creating the feature array
    X = preprocessing.scale(X) #processing the feature array
    X_lately = X[-forecast_out:] #creating the column i want to use later in the predicting method
    X = X[:-forecast_out] # X that will contain the training and testing
    label.dropna(inplace=True) #dropping na values
    y = np.array(label)  # assigning Y
    X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=test_size, random_state=0) #cross validation

    response = [X_train,X_test , Y_train, Y_test , X_lately]
    return response
```

Ma'lumotlarni tayyorlash
Yuqoridagi bo'limda men python kutubxonalarini import qilish orqali aksiyalar narxini bashorat qilish vazifasini boshladim. Endi men uni chiziqli regressiya modeliga osongina moslashimiz uchun ma'lumotlar to'plamini tayyorlaydigan funktsiyani yozaman:

```python
df = pd.read_csv("prices.csv")
df = df[df.symbol == "GOOG"]
```


Yuqoridagi funktsiyani osongina tushunishingiz mumkin, chunki men har bir satrning ishlashini bosqichma-bosqich aytib berdim. Endi qilish kerak bo'lgan keyingi narsa ma'lumotlarni o'qishdir:

```python
forecast_col = 'close'
forecast_out = 5
test_size = 0.2
```

Endi biz yuqoridagi bo'limda yaratilgan funktsiyada allaqachon tayyorlangan uchta kirish o'zgaruvchisini tayyorlashimiz kerak. Biz qaysi ustunni bashorat qilmoqchi ekanligimizni eslatuvchi kirish o'zgaruvchisini e'lon qilishimiz kerak. Biz e'lon qilishimiz kerak bo'lgan keyingi o'zgaruvchi bu qancha masofani taxmin qilishni xohlashimizdir.

Va biz e'lon qilishimiz kerak bo'lgan oxirgi o'zgaruvchi - bu test to'plamining o'lchami qancha bo'lishi kerak. Endi barcha o'zgaruvchilarni e'lon qilamiz:

```python
X_train, X_test, Y_train, Y_test , X_lately =prepare_data(df,forecast_col,forecast_out,test_size); #calling the method were the cross validation and data preperation is in
learner = LinearRegression() #initializing linear regression model

learner.fit(X_train,Y_train) #training the linear regression model
```

Qimmatli qog'ozlar narxini bashorat qilish uchun mashinani o'rganishni qo'llash
Endi men ma'lumotlarni ajrataman va chiziqli regressiya modeliga moslashtiraman:

```python
score=learner.score(X_test,Y_test)#testing the linear regression model
forecast= learner.predict(X_lately) #set that will contain the forecasted data
response={}#creting json object
response['test_score']=score
response['forecast_set']=forecast

print(response)
```
